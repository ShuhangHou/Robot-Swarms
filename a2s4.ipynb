```python
%pylab 
Turn = {0:"right",1:"up",2:"left",3:"down",4:"wait"} # 2D
K = 5
size = 4
gama = 1
reward = 0
thita = 0.01
E = 0.1
np.random.seed(0)
def getValue(pos1,boom1,pos2,boom2):
    if boom1:
        value1 = -2
    else:
        value1 = -1
    if boom2:
        value2 = -2
    else:
        value2 = -1
    # goal
    
    if pos1 == 0 or pos1 == (size*size-1):
        value1 = 0
    if pos2 == 0 or pos2 == (size*size-1):
        value2 = 0
    
    # crash 
    if pos1 == pos2:
        if pos1 != 0 and pos1 != (size*size -1) and pos2 != 0 and pos2 != (size*size -1):
            value2 = -2
    return value1,value2

def move(x,d):
    boom = False
    
    #if x == 0 or x == (size*size -1):
    #    return x,boom
    #else:
    if d == 1:
        if x > size - 1:
            x = x - size
        else:
            boom = True      
    elif d == 3:
        if x < size * size - size:
            x = x + size
        else:
            boom = True   
    elif d == 0:
        if ((x+1) % size) > 0:
            x = x + 1
        else:
            boom = True   
    elif d == 2:
        if (x % size) > 0 :
            x = x - 1
        else:
            boom = True   
    return x,boom
                
    
# init grid
grid = list(range(size*size))

Q = [[[0 for i in range(K*K)] for i in range(size*size)] for i in range(size*size)]
V = [[0 for i in range(size*size)] for i in range(size*size)]
#update_time = [[0 for i in range(K)] for i in range(size*size)]


#using random policy
Vk = V
delta = 1
t = 0
total_reward = 0
# 2 robots
#while (delta > thita):
while (t<5):
    delta = 0
    # robot 1
    for i1 in range(1,size*size-1): 
        # robot 2
        for i2 in range(1,size*size-1): 
            for j in range(K*K):
                m1,m2 = divmod(j, K)
                pos1,boom1 = move(i1,m1)
                pos2,boom2 = move(i2,m2)
                value1,value2 = getValue(pos1,boom1,pos2,boom2)
                reward = 0.5 * value1  + 0.5 * value2 + gama*V[pos1][pos2]
                Q[i1][i2][j] = reward 
    for i1 in range(size*size):
        for i2 in range(size*size):
            for j in range(K*K):
                total_reward += Q[i1][i2][j]/25 
            v = V[i1][i2]
            Vk[i1][i2] = total_reward 
            total_reward = 0
            delta = max(delta,abs(v-Vk[i1][i2]))
    V = Vk
    print(Vk)
    t += 1
        
solution = []
for i1 in range(size*size):
    for i2 in range(size*size):
        solution.append(argmax(Q[i1][i2]))
print('times = ',t)
print(solution)
```
